```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('processed-data.csv')
df
```

```python
# dropping tables and renaming columns
df = df.drop(columns=['Severity_Mild', 'Severity_Moderate','None_Experiencing','None_Sympton'])
df.rename(columns={'Severity_None': 'Target'}, inplace=True)
```

```python
df
```

```python
x = df.drop(columns=['Target'])
y  = df['Target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42, class_weight='balanced')
model.fit(x_train, y_train)
y_hat = model.predict(x_test)
y_hat
```

```python
print('**** ACCURACY_SCORE **** \n\n', accuracy_score(y_test, y_hat), '\n')
print('**** CLASSIFICATION_REPORT **** \n\n', classification_report(y_test, y_hat), '\n')
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 2, 5, 7, 10, 12, 15, 17],
    'min_samples_split': [2, 5, 10]
}

# Initialize Grid Search
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit Grid Search
grid_search.fit(x_train, y_train)

# Best parameters
best_params = grid_search.best_params_
print(f'Best parameters: {best_params}')
```

```python
# Initialize the model with best parameters
final_model = RandomForestClassifier(**best_params, random_state=42, class_weight='balanced')

# Train the model
final_model.fit(x_train, y_train)
```

```python
import joblib

# Save the model
joblib.dump(final_model, 'final_model.joblib')

# Load the model (for future predictions)
loaded_model = joblib.load('final_model.joblib')
```

```python
from tkinter import *
from tkinter import ttk
import pandas as pd
import joblib

# Load the trained model
loaded_model = joblib.load('final_model.joblib')

def predict_asthma():
    # Preprocess input data
    symptoms = {
        "Tiredness": tiredness_var.get(),
        "Dry-Cough": dry_cough_var.get(),
        "Difficulty-in-Breathing": breathing_difficulty_var.get(),
        "Sore-Throat": sore_throat_var.get(),
        "Pains": pains_var.get(),
        "Nasal-Congestion": nasal_congestion_var.get(),
        "Runny-Nose": runny_nose_var.get(),
    }

    selected_age = age_var.get()
    selected_gender = gender_var.get()

    age_encoding = [1 if selected_age == age_option else 0 for age_option in age_options]
    gender_encoding = [1 if selected_gender == gender_option else 0 for gender_option in gender_options]

    data = {
        'Tiredness': [symptoms["Tiredness"]],
        'Dry-Cough': [symptoms["Dry-Cough"]],
        'Difficulty-in-Breathing': [symptoms["Difficulty-in-Breathing"]],
        'Sore-Throat': [symptoms["Sore-Throat"]],
        'Pains': [symptoms["Pains"]],
        'Nasal-Congestion': [symptoms["Nasal-Congestion"]],
        'Runny-Nose': [symptoms["Runny-Nose"]],
        'Age_0-9': age_encoding[0],
        'Age_10-19': age_encoding[1],
        'Age_20-24': age_encoding[2],
        'Age_25-59': age_encoding[3],
        'Age_60+': age_encoding[4],
        'Gender_Female': gender_encoding[0],
        'Gender_Male': gender_encoding[1]
    }

    input_df = pd.DataFrame(data)

    # Make prediction
    prediction = loaded_model.predict(input_df)

    # Display result
    result_label.config(text=f"Prediction: {'Asthma' if prediction[0] == 0 else 'Not Asthma'}\n Prediction value: {prediction}" )
    #result_label.config(text=f"Prediction: {'Asthma' if prediction[0] == 1 else 'Not Asthma'}")

root = Tk()
root.title("Asthma Detection")

# Menu Bar
menu_bar = Menu(root)
root.config(menu=menu_bar)

# File Menu
file_menu = Menu(menu_bar, tearoff=0)
menu_bar.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="Exit", command=root.quit)

# Symptom Frame
symptom_frame = ttk.Frame(root)
symptom_frame.pack(padx=10, pady=10, anchor="w")

tiredness_var = BooleanVar()
dry_cough_var = BooleanVar()
breathing_difficulty_var = BooleanVar()
sore_throat_var = BooleanVar()
pains_var = BooleanVar()
nasal_congestion_var = BooleanVar()
runny_nose_var = BooleanVar()

symptom_vars = [
    tiredness_var, dry_cough_var, breathing_difficulty_var, sore_throat_var, pains_var, nasal_congestion_var, runny_nose_var
]

symptom_labels = [
    "Tiredness", "Dry-Cough", "Difficulty-in-Breathing", "Sore-Throat",
     "Pains", "Nasal-Congestion", "Runny-Nose"
]

for index, label in enumerate(symptom_labels):
    checkbox = ttk.Checkbutton(symptom_frame, text=label, variable=symptom_vars[index])
    checkbox.pack(anchor="w")

# Age Frame
age_frame = ttk.Frame(root)
age_frame.pack(padx=10, pady=10, anchor="w")

age_var = StringVar()
age_var.set("Age_0-9")  # Default selection

age_options = [
    "Age_0-9", "Age_10-19", "Age_20-24", "Age_25-59", "Age_60+"
]

for age_option in age_options:
    rb = ttk.Radiobutton(age_frame, text=age_option, variable=age_var, value=age_option)
    rb.pack(anchor="w")

# Gender Frame
gender_frame = ttk.Frame(root)
gender_frame.pack(padx=10, pady=10, anchor="w")

gender_var = StringVar()
gender_var.set("Female")  # Default selection

gender_options = [
    "Female", "Male"
]

for gender_option in gender_options:
    rb = ttk.Radiobutton(gender_frame, text=gender_option, variable=gender_var, value=gender_option)
    rb.pack(anchor="w")

# Result Label
result_label = ttk.Label(root, text="", style="Result.TLabel")
result_label.pack(pady=10)

# Submit Button
submit_button = ttk.Button(root, text="Predict", command=predict_asthma)
submit_button.pack()

root.mainloop()
```
